```python
import quandl
import datetime
from datetime import date
import pandas as pd
```

```python
WTI_monthly = quandl.get("FRED/DCOILWTICO",start_date="1990-01-01", end_date="2020-06-01", authtoken="TYxF4cUU1kEsRwg8QEdu", collapse="monthly", order="asc")
print(WTI_monthly)
```

```python
BRENT_monthly = quandl.get("FRED/DCOILBRENTEU", start_date="1990-01-01", end_date="2020-06-01",  authtoken="TYxF4cUU1kEsRwg8QEdu", collapse="monthly", order="asc")
print(BRENT_monthly)
```

```python
##ARAB_LIGHT_monthly = quandl.get("EIA/PET_ISA4990008_M", start_date="1990-01-01", end_date="2020-06-01",  authtoken="TYxF4cUU1kEsRwg8QEdu", collapse="monthly", order="asc")
## print(ARAB_LIGHT_monthly)
```

```python
# https://www.quandl.com/data/EIA/STEO_COPRPUS_M-U-S-Crude-Oil-Production-Monthly
US_oil_prod_monthly = quandl.get("EIA/STEO_COPRPUS_M", start_date="1990-01-01", end_date="2020-06-01",  authtoken="TYxF4cUU1kEsRwg8QEdu", collapse="monthly", order="asc")
print(US_oil_prod_monthly)
```

```python
# World Fuel Consumption https://www.quandl.com/data/EIA/STEO_PATC_WORLD_M-Liquid-Fuels-Consumption-Total-World-Monthly
world_fuel_cons_monthly = quandl.get("EIA/STEO_COPRPUS_M", start_date="1990-01-01", end_date="2020-06-01",  authtoken="TYxF4cUU1kEsRwg8QEdu", collapse="monthly", order="asc")
print(world_fuel_cons_monthly)
```

```python
## FYI found a quandl for USA fuel oil consuption so added it here - not sure if we want to keep it consistent with USA oil prduction and consumption
# US Fuel Oil Consumption, Monthly - https://www.quandl.com/data/EIA/STEO_D2TCPUS_M-Fuel-Oil-Consumption-Monthly
US_fuel_cons_monthly = quandl.get("EIA/STEO_D2TCPUS_M", start_date="1990-01-01", end_date="2020-06-01",  authtoken="TYxF4cUU1kEsRwg8QEdu", collapse="monthly", order="asc")
print(US_fuel_cons_monthly)
```

```python
oil_data=WTI_monthly.merge(BRENT_monthly,how='outer', left_index=True, right_index=True)
oil_data=oil_data.rename(columns={"Value_x": "WTI", "Value_y": "BRENT"})
oil_data
```

```python
oil_data=oil_data.merge(US_oil_prod_monthly,how='outer', left_index=True, right_index=True)
oil_data=oil_data.rename(columns={"Value": "US_PROD"})
oil_data
```

```python
oil_data=oil_data.merge(US_fuel_cons_monthly,how='outer', left_index=True, right_index=True)
oil_data=oil_data.rename(columns={"Value": "US_CONS"})
oil_data
```

```python
oil_data=oil_data.merge(world_fuel_cons_monthly,how='outer', left_index=True, right_index=True)
oil_data=oil_data.rename(columns={"Value": "WORLD_CONS"})
oil_data
```

```python
oil_data.dropna(inplace=True)
oil_data
```

```python
### export to CSV
oil_data.to_csv("oil_data.csv")
```
