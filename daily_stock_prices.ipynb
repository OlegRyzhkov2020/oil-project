```python
daily_data.dropna(inplace=True)
daily_data
```

# Create Lagged Variables of Daily Change in Stock Prices & Change in Oil Prices

```python
daily_data['WTI_lag'] = daily_data['WTI'].shift(1)
daily_data['WTI_change'] = daily_data['WTI']-daily_data['WTI_lag']
daily_data['WTI_increase'] = 1 * (daily_data['WTI_change']  > 0)
daily_data['WTI_decrease'] = 1 * (daily_data['WTI_change']  < 0)
daily_data['BRENT_lag'] = daily_data['BRENT'].shift(1)
daily_data['BRENT_change'] = daily_data['BRENT']-daily_data['BRENT_lag']
daily_data['BRENT_increase'] = 1 * (daily_data['BRENT_change']  > 0)
daily_data['BRENT_decrease'] = 1 * (daily_data['BRENT_change']  < 0)
daily_data['exxon_lag'] = daily_data['exxon_close'].shift(1)
daily_data['exxon_change'] = daily_data['exxon_close']-daily_data['exxon_lag']
daily_data['exxon_increase'] = 1 * (daily_data['exxon_change']  > 0)
daily_data['exxon_decrease'] = 1 * (daily_data['exxon_change']  < 0)
daily_data['chevron_lag'] = daily_data['chevron_close'].shift(1)
daily_data['chevron_change'] = daily_data['chevron_close']-daily_data['chevron_lag']
daily_data['chevron_increase'] = 1 * (daily_data['chevron_change']  > 0)
daily_data['chevron_decrease'] = 1 * (daily_data['chevron_change']  < 0)
daily_data['conoco_lag'] = daily_data['conoco_close'].shift(1)
daily_data['conoco_change'] = daily_data['conoco_close']-daily_data['conoco_lag']
daily_data['conoco_increase'] = 1 * (daily_data['conoco_change']  > 0)
daily_data['conoco_decrease'] = 1 * (daily_data['conoco_change']  < 0)
daily_data['eog_lag'] = daily_data['eog_close'].shift(1)
daily_data['eog_change'] = daily_data['eog_close']-daily_data['eog_lag']
daily_data['eog_increase'] = 1 * (daily_data['eog_change']  > 0)
daily_data['eog_decrease'] = 1 * (daily_data['eog_change']  < 0)
daily_data['valero_lag'] = daily_data['valero_close'].shift(1)
daily_data['valero_change'] = daily_data['valero_close']-daily_data['valero_lag']
daily_data['valero_increase'] = 1 * (daily_data['valero_change']  > 0)
daily_data['valero_decrease'] = 1 * (daily_data['valero_change']  < 0)
daily_data['baker_lag'] = daily_data['baker_close'].shift(1)
daily_data['baker_change'] = daily_data['baker_close']-daily_data['baker_lag']
daily_data['baker_increase'] = 1 * (daily_data['baker_change']  > 0)
daily_data['baker_decrease'] = 1 * (daily_data['baker_change']  < 0)
daily_data
```

```python
daily_data=daily_data.drop(columns=['WTI_lag','WTI_change','BRENT_lag','BRENT_change','exxon_lag', 'exxon_change', 'chevron_lag', 'chevron_change','conoco_lag','conoco_change','eog_lag','eog_change','valero_lag','valero_change','baker_lag','baker_change'])
daily_data
```

```python
### export to CSV
daily_data.to_csv("daily_data.csv")
```
