```python
# padd = "PADD 1"
def clean_df(df,padd,padd_name):
    # correct the columns names
    df=df.rename(columns={df.columns[1] :"total"})
    i=0
    while i< len(df.columns):
        df=df.rename(columns={
            df.columns[i] : df.columns[i].replace(f"{padd_name} ({padd}) Exports to ","").replace(" of Crude Oil (Thousand Barrels)","")
            })
        i +=1
        
    # replace NaN by 0
    df.fillna(0, inplace=True)
    
    # get ohter countries colum and value
    i=2
    df["other"] = df.total-df.iloc[:, 2:(len(df.columns))].sum(1) 
    
    # insert padd column as first one
    df.insert(0,"padd", padd.replace(" ",""))
    
    # insert year column and then delete Date column
    df.insert(1,"year", df["Date"].dt.year)
    df.drop("Date", axis=1, inplace=True)
    
    # cast all numbers as int
    i=3
    while i < len(df.columns):
        df = df.astype({df.columns[i]: "int"})
        i +=1
    
    #select only 2019 year
    df = df[df["year"]==2019]
    
    # reset index
#     df.reset_index(drop=True)
    
    
    # return df.head
    return  df.reset_index(drop=True)
```

# save data in geojson format
 - inspiration/source: https://geoffboeing.com/2015/10/exporting-python-data-geojson/

```python
def df_to_geojson(df, properties, lon1="start_lon", lat1="start_lat", lon2="end_lon", lat2="end_lat",):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'LineString',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [
                                    [row[lon1],row[lat1]],
                                    [row[lon2],row[lat2]]
                                    ]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson
```

```python
properties = ['location', 'padd', 'volume','code']
geojson_dict = df_to_geojson(import_df, properties)
geojson_str = json.dumps(geojson_dict, indent=2)
```

```python
output_filename = 'import.geojson'
with open(output_filename, 'w') as output_file:
    output_file.write(format(geojson_str))
    
# how many features did we save to the geojson file?
print('{} geotagged features saved to file'.format(len(geojson_dict['features'])))
```
