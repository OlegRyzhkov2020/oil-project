```python
import pandas as pd
import numpy as np
import os
```

```python
gdp_df = pd.read_csv(os.path.join("data", "FRED-GDP.csv"))
gdp_df
```

```python
gdp_df["year"]=gdp_df["Date"].str.split("-", 2).str.get(0)
gdp_df["month"]=gdp_df["Date"].str.split("-", 2).str.get(1)
gdp_df["quater"]="Q1"
```

```python
gdp_df.loc[gdp_df["month"]=="04","quater"]="Q2"
gdp_df.loc[gdp_df["month"]=="07","quater"]="Q3"
gdp_df.loc[gdp_df["month"]=="12","quater"]="Q4"
```

```python
gdp_df.head()
```

```python
gdp_df.drop("Date", axis=1, inplace=True)
gdp_df.drop("month", axis=1, inplace=True)
```

```python
gdp_df=gdp_df.rename(columns={"Value" :"gdp"})
```

```python
gdp_df
```

```python
export_df=pd.read_csv(os.path.join("data", "U.S._Exports_of_Crude_Oil.csv"), header=4)
export_df
```

```python
export_df.info()
```

```python
month_year=export_df["Month"]
# months["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
export_df["year"]=export_df["Month"].str.split(" ", 1).str.get(1)
export_df["Month"]=export_df["Month"].str.split(" ", 1).str.get(0)
export_df.insert(0,"quater", "Q1")
```

```python
export_df.loc[export_df["Month"]=="Apr","quater"]="Q2"
export_df.loc[export_df["Month"]=="May","quater"]="Q2"
export_df.loc[export_df["Month"]=="Jun","quater"]="Q2"
export_df.loc[export_df["Month"]=="Jul","quater"]="Q3"
export_df.loc[export_df["Month"]=="Aug","quater"]="Q3"
export_df.loc[export_df["Month"]=="Sep","quater"]="Q3"
export_df.loc[export_df["Month"]=="Oct","quater"]="Q4"
export_df.loc[export_df["Month"]=="Nov","quater"]="Q4"
export_df.loc[export_df["Month"]=="Dec","quater"]="Q4"
```

```python
export_df
```

```python
export_byQ=export_df.groupby(["year", "quater"]).sum()
export_byQ.reset_index(inplace=True)
export_byQ=export_byQ.rename(columns={"U.S. Exports of Crude Oil Thousand Barrels" :"export"})
```

```python
df=pd.merge(gdp_df, export_byQ, how="outer")
df.dropna(inplace=True)
df
```
